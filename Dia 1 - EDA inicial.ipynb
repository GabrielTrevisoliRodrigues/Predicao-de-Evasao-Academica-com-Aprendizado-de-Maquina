A partir da EDA inicial a baixo podemos concluir e ressaltar alguns pontos importantes:

```python
#bibliotecas que estaremos usando nessa etapa inicial
import pandas as pd
import numpy as np

#importando o banco de dados
#obs: ele é um arquivo csv 

df = pd.read_csv("../DataSet/data.csv", sep=";")

#vizualização inicial
print("Visualização inicial:")
display(df.head())
print()

print("Dimensões do dataset:")
print(df.shape)
print()

#informações gerais 
print("Informações gerais do DataFrame:")
df.info()
print()

#estatisticas descritivas
print("Estatisticas descritivas")
display(df.describe(include='all'))
print()

#valores ausentes
print("valores ausentes por coluna")
print(df.isnull().sum())
```

MARKDOWN CÉLULA 1:

Não existem valores faltantes em nenhuma coluna, o que nos permite economizar tempo na prepração e limpeza de dados. 
Também é bom ressaltar que todas os atributos são de tipos númericos int ou float, exceto pelo atributo target do tipo object que nos informa sobre a classificação de cada aluno (dropout, enrolled, and graduate).




................................................................................................................................................................................................................................................................





```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(data=df, x='Target')
plt.title("Distribuição dos estados dos alunos (evasão, conclusão, cursando)")
plt.show()


distribuicao = pd.DataFrame({
    'Contagem': df['Target'].value_counts(),
    'Percentual (%)': df['Target'].value_counts(normalize=True) * 100
})
display(distribuicao)
```

MARKDOWN CÉLULA 2:
Podemos observar um leve desbalanceamento de nas classes sendo a minoritaria apenas 17%, entretanto essa diferença aparenta ser leve demais para justificar a aplicação de tecnicas de balanceamento.
